```python
import cv2

# Load the image
image_path = '/content/vincard_empty.jpg'
image = cv2.imread(image_path)

# List of expected text labels
labels = ['TYP', 'VIN', 'GVW', 'GCW', 'AXLE 1', 'AXLE 2', 'AXLE 3', 'AXLE 4', 'AXLE 5']

# Initialize variables
positions = {}
current_label = 0
refPt = []
cropping = False

def click_and_crop(event, x, y, flags, param):
    global refPt, cropping, current_label

    if event == cv2.EVENT_LBUTTONDOWN:
        refPt = [(x, y)]
        cropping = True

    elif event == cv2.EVENT_LBUTTONUP:
        refPt.append((x, y))
        cropping = False

        cv2.rectangle(image, refPt[0], refPt[1], (0, 255, 0), 2)
        cv2.putText(image, labels[current_label], (refPt[0][0], refPt[0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        cv2.imshow("image", image)

        # Calculate the width and height
        x1, y1 = refPt[0]
        x2, y2 = refPt[1]
        width = x2 - x1
        height = y2 - y1

        # Save the position
        positions[labels[current_label]] = (x1, y1, width, height)
        current_label += 1

        # Check if all labels are processed
        if current_label >= len(labels):
            print("Positions:", positions)
            cv2.destroyAllWindows()

# Clone the image to display
clone = image.copy()
cv2.namedWindow("image")
cv2.setMouseCallback("image", click_and_crop)

while True:
    cv2.imshow("image", image)
    key = cv2.waitKey(1) & 0xFF

    if key == ord("q") or current_label >= len(labels):
        break

cv2.destroyAllWindows()
```
